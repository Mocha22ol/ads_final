```python
import pandas as pd
import numpy as np

import requests
import re
import json
from pandas.io.json import json_normalize
import math

import warnings
warnings.filterwarnings('ignore')
```

```python
api_key = ""

data = pd.read_csv('ads_zillow_clean.csv')
data.head()
```

```python
data['distance_ts'] = 'TBD'
data['duration_ts'] = 'TBD'
data['distance_db'] = 'TBD'
data['duration_db'] = 'TBD'
data['distance_mn'] = 'TBD'
data['duration_mn'] = 'TBD'
data.head()
```

```python
dest_list = ["Times Sq-42 St Station, New York, NY", "City Hall, New York, NY","Downtown Brooklyn, Brooklyn, NY"]
for i in range(5574, len(data)):
    for dest in dest_list:
        origin_address = data.clean[i]
      
        url = ('https://maps.googleapis.com/maps/api/distancematrix/json?units=imperial&origins={}&destinations={}&mode=transit&key={}'.format(origin_address, dest, api_key))
        #print(url)
        req = requests.get(url)
        res = req.json()
        res = json.dumps(res)
        res_new = eval(res)

        distance = str(eval(str(eval(str(eval(str(res_new['rows'])[1:-1])['elements'])[1:-1])['distance']))['text'])
        duration = str(eval(str(eval(str(eval(str(res_new['rows'])[1:-1])['elements'])[1:-1])['duration']))['text'])
    
        if (dest == "Times Sq-42 St Station, New York, NY"):
            data['distance_ts'][i] = distance
            data['duration_ts'][i] = duration
        elif (dest == "Downtown Brooklyn, Brooklyn, NY"):
            data['distance_db'][i] = distance
            data['duration_db'][i] = duration
        elif (dest == "City Hall, New York, NY"):
            data['distance_mn'][i] = distance
            data['duration_mn'][i] = duration
    print(i)
```

```python
data.to_csv("zillow_with_travel_time.csv")
```
